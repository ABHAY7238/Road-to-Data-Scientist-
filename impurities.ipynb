**Entropy**

```python
import math
```

```python
#now make the data set
#suppose we have the dataset of two subsets a and b and both have 10 elements 4 of them in class A and 6 in class B
n_a = 4
n_b = 6
total = n_a + n_b
```

```python
'now lets calculate the proportions of each class'
p_a = n_a/total
p_b = n_b/total
'now print the prporstions'
print(f'proporstions of a = ',p_a)
print(f'porporstions of b = ',p_b)
```

```python
'gini impurity'
gini = 1 - (p_a**2 + p_b**2)
print(f'gini_impurity = ',gini)
```

*Information gain impurities* <br>
Information gain is the difference between the impurity of the starting node and the weighted average of the impurity of the child nodes.MM

```python
'assuming a split on some features divides the dataset into two subsets'
'now lets calculate the impurity of each subset'
'subset 1 : 2 elements in class A and 3 in class B'
'subset 2 : 2 elements in class A and 3 in class B'
n_1_A , n_1_B = 2,3
n_2_A , n_2_B = 2,3
total_1 = n_1_A + n_1_B
total_2 = n_2_A + n_2_B
'now calculate the proportions of each class'
p_1_A = n_1_A/total_1
p_1_B = n_1_B/total_1
'now calculate the entropy of 1st subset'
entropy_1 = -p_1_A*math.log2(p_1_A) - p_1_B*math.log2(p_1_B)
'now calculate the proportions of each class'
p_2_A = n_2_A/total_2
p_2_B = n_2_B/total_2
'now calculate the entropy of 2nd subset'
entropy_2 = -p_2_A*math.log2(p_2_A) - p_2_B*math.log2(p_2_B)
'now calculate the information gain'
info_gain = gini - (total_1/total)*entropy_1 - (total_2/total)*entropy_2
print(f'information gain = ',info_gain)
```

**Decision Tree examples**

```python
#import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
```

```python
#load the dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#drop the deck column from the dataset
df.drop('deck',axis=1,inplace=True)
```

```python
#now impute the null values of 'age' and 'fare' by using SimpleImputer
imputer = SimpleImputer(strategy='median')
df['age'] = imputer.fit_transform(df[['age']])
df['fare'] = imputer.fit_transform(df[['fare']])
```

```python
#now impute the embarked and embarked_town by using mode
imputer = SimpleImputer(strategy = 'most_frequent')
df[['embarked']] = imputer.fit_transform(df[['embarked']])
df[['embark_town']] = imputer.fit_transform(df[['embark_town']])
```

```python
df.isnull().sum()
```

```python
#now enode the categorical and object data using the labelEncoder
le = LabelEncoder()
df['sex'] = le.fit_transform(df['sex'])
df['embarked'] = le.fit_transform(df['embarked'])
df['embark_town'] = le.fit_transform(df['embark_town'])
df['who'] = le.fit_transform(df['who'])
df['alive'] = le.fit_transform(df['alive'])
df['class'] = le.fit_transform(df['class'])

```

```python
df.head()
```

```python
#split the data into x and y
x = df.drop('survived',axis=1)
y = df['survived']
```

```python
#split the data into train and test
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#now predict the model and train the model
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
#predict the model
y_pred = model.predict(x_test)

```

```python
#evaluate the model
accuracy = accuracy_score(y_test,y_pred)
print(f'accuracy_score = ',accuracy_score(y_test,y_pred))
print(f'confusion_matrix =',confusion_matrix(y_test,y_pred))
print(f'classification_report =',classification_report(y_test,y_pred))
```

```python
#save the decision tree classifier
import pickle
with open('decision_tree_classifier.pkl','wb') as f:
    pickle.dump(model,f)
```
